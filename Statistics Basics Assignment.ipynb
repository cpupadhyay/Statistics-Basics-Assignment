# Theoretical Questions

1. What is statistics, and why is it important?
   -> Statistics is the science of collecting, analyzing, interpreting, and presenting data. It is important because it helps in making informed decisions       based on data-driven insights.

2. What are the two main types of statistics?
   -> The two main types of statistics are descriptive statistics and inferential statistics.

3. What are descriptive statistics?
   -> Descriptive statistics summarize and organize data (e.g., mean, median, charts).

4. What is inferential statistics?
   -> Inferential statistics involves using data from a sample to make estimates, predictions, or generalizations about a larger population. It helps         assess relationships, test hypotheses, and determine the likelihood that an observed pattern is due to chance.

5. What is sampling in statistics?
   -> Sampling in statistics is the process of selecting a subset (sample) from a larger group (population) to gather data and make inferences about          the entire population. It allows researchers to study a manageable amount of data while still drawing meaningful conclusions.

6. What are the different types of sampling methods?
   -> The main types of sampling methods in statistics are:
      Random Sampling, Systematic Sampling, Statified Sampling, Cluster Sampling, Convenience Sampling, Quota Sampling. 

7. What is the difference between random and non-random sampling?
   -> The difference between random and non-random sampling lies in how the sample is selected:
      -> Random Sampling: Every individual in the population has an equal and known chance of being selected. It reduces bias and supports                       generalization to the population (e.g., simple random sampling, stratified sampling).
      -> Non-Random Sampling: Individuals are selected based on non-random criteria (like convenience or judgment), so not everyone has an equal                 chance. It's quicker but may introduce bias (e.g., convenience sampling, quota sampling).

8. Define and give examples of qualitative and quantitative data.
   -> Qualitative Data: Non-numerical information that describes qualities or categories.
      -> Example: Colors of cars (red, blue, black), types of cuisine (Italian, Chinese), customer feedback (satisfied, unsatisfied).
   -> Quantitative Data: Numerical information that can be measured or counted.
      -> Example: Age of people (23, 35, 42), number of products sold (100, 250), height in cm (160, 175).

9. What are the different types of data in statistics?
   -> Quantitative Data and Qualitative Data 

10. Explain nominal, ordinal, interval, and ratio levels of measurement.
    -> Nominal (Qualitative): Categorizes data without any order.
    -> Ordinal (Qualitative): Categorizes data with a meaningful order, but intervals are not equal.
    -> Interval (Quantitative): Numeric data with equal intervals, but no true zero.
    -> Ratio (Quantitative) : Numeric data with equal intervals and a true zero point.

11. What is the measure of central tendency?
    -> The measure of central tendency are: Mean,Median,Mode

12. Define mean, median, and mode.
    -> Here’s a clear definition of mean, median, and mode:
       -> Mean : The average of all values in a dataset.
       -> Median : The middle value of a dataset when it is arranged in ascending or descending order.
       -> Mode : The value that appears most frequently in a dataset.

13. What is the significance of the measure of central tendency?
    -> The measure of central tendency provides a single value that summarizes the center of a dataset, helping to identify patterns or typical values.        It is crucial for comparing different datasets and making data-driven decisions.

14. What is variance, and how is it calculated?
    -> Variance is a measure of how much the values in a dataset differ from the mean. It quantifies the spread or dispersion of the data points.
    And Steps to calculate variance
    -> Find the mean of the data.
    -> Subtract the mean from each data point.
    -> Square each result from step 2.
    -> For population variance, average the squared results. For sample variance, divide by n-1.

15. What is standard deviation, and why is it important?
    -> Standard deviation is the square root of variance and measures the average distance of data points from the mean, indicating the data's spread.         It is important because it provides a more interpretable measure of variability in the same units as the data itself.


16. Define and explain the term range in statistics.
    -> Range in statistics is the difference between the highest and lowest values in a dataset. It provides a simple measure of the spread or                 dispersion of the data.
       Formula:Range = Maximum value − Minimum value

17. What is the difference between variance and standard deviation?
    -> Variance measures the average squared deviation from the mean, indicating the spread of data.Standard deviation is the square root of                   variance and provides a more interpretable measure of spread in the same units as the data. Essentially, standard deviation is the "average             distance" from the mean, while variance gives a squared value.


18. What is skewness in a dataset?
    -> Skewnes refers to the asymmetry or lack of symmetry in a dataset's distribution. It indicates whether the data is skewed to the left (negative          skew) or to the right (positive skew).
    -> Positive skew (right skew) : The right tail of the distribution is longer, and most values are concentrated on the left.
    -> Negative skew (left skew) : The left tail of the distribution is longer, and most values are concentrated on the right.
    -> Zero skew : The distribution is symmetric, like a normal distribution.

19. What does it mean if a dataset is positively or negatively skewed?
    -> Positively Skewed (Right Skewed) : If a dataset is positively skewed, it means the right tail (larger values) is longer, and most of the data           points are concentrated on the left side. The mean is typically greater than the median in this case.
    -> Negatively Skewed (Left Skewed) : If a dataset is negatively skewed, the left tail (smaller values) is longer, and most of the data points are          concentrated on the right side. The mean is usually less than the median.

20. Define and explain kurtosis.
    -> Kurtosis is a statistical measure that describes the shape of a distribution's tails and the peakness of the dataset compared to a normal               distribution. It provides insight into the outlier behavior of the data.
    -> Types of Kurtosis:
       -> Leptokurtic (Positive Kurtosis): The distribution has heavy tails and a sharp peak, indicating a higher likelihood of outliers.
          -> Example: Stock market returns, where extreme events (outliers) are more likely.
       -> Platykurtic (Negative Kurtosis): The distribution has light tails and a flatter peak, suggesting fewer outliers.
          -> Example : Uniform distribution, where all values are fairly close to each other.
       -> Mesokurtic (Zero Kurtosis): The distribution has a normal shape (similar to a normal distribution with a kurtosis value of 3).
          -> Example : A normal distribution, where tail behavior is moderate.
       Kurtosis helps identify the presence of outliers and extreme values in a dataset, indicating whether data points are more or less concentrated          in the tails compared to a normal distribution.

21. What is the purpose of covariance?
    -> Covariance is a measure of the relationship between two variables and indicates how they change together. Specifically, it shows whether an             increase in one variable leads to an increase or decrease in the other.

22. What does correlation measure in statistics?
    -> Correlation measures the strength and direction of the linear relationship between two variables. It quantifies how closely the variables move          together.

```python
#5. Write a Python function to calculate the range of a dataset.
#code
def calculate_range(data):
    # Ensure the dataset is not empty
    if len(data) == 0:
        return "Dataset is empty"
    
    # Calculate the range
    data_range = max(data) - min(data)
    return data_range

# Example dataset
data = [3, 5, 7, 9, 1, 3, 5, 7]

# Call the function and display the result
range_value = calculate_range(data)
print("Range of the dataset:", range_value)
```

```python
#12. Implement and compare simple random sampling and systematic sampling.
#code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample population dataset
np.random.seed(42)
population = pd.DataFrame({
    'ID': np.arange(1, 101),
    'Value': np.random.randint(10, 100, size=100)
})

# Simple Random Sampling
simple_random_sample = population.sample(n=10, random_state=1)

# Systematic Sampling
k = len(population) // 10  # Sampling interval
start = np.random.randint(0, k)
systematic_sample = population.iloc[start::k][:10]

# Plotting the samples for comparison
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.bar(simple_random_sample['ID'], simple_random_sample['Value'], color='skyblue')
plt.title('Simple Random Sampling')
plt.xlabel('ID')
plt.ylabel('Value')

plt.subplot(1, 2, 2)
plt.bar(systematic_sample['ID'], systematic_sample['Value'], color='salmon')
plt.title('Systematic Sampling')
plt.xlabel('ID')
plt.ylabel('Value')

plt.tight_layout()
plt.show()

# Optional: Print the samples
print("Simple Random Sample:\n", simple_random_sample)
print("\nSystematic Sample:\n", systematic_sample)
```

```python
#20. Write a Python program to visualize covariance using a heatmap.
#code
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Simulate multivariate data
np.random.seed(0)
x = np.random.normal(50, 10, 100)
y = 2 * x + np.random.normal(0, 5, 100)  # Strong positive covariance with x
z = np.random.normal(50, 10, 100)        # Unrelated variable

# Step 2: Create a DataFrame
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Z': z
})

# Step 3: Compute covariance matrix
cov_matrix = df.cov()

# Step 4: Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Covariance Matrix Heatmap')
plt.show()
```

```python
#24. Implement the Pearson and Spearman correlation coefficients for a dataset.
#code
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Step 1: Simulate a dataset
np.random.seed(0)
x = np.random.normal(50, 10, 1000)  # Normally distributed variable
y = 2 * x + np.random.normal(0, 5, 1000)  # Linearly related to x
z = np.random.normal(50, 10, 1000)  # Unrelated variable

# Step 2: Create a DataFrame
df = pd.DataFrame({'X': x, 'Y': y, 'Z': z})

# Step 3: Compute Pearson correlation between X and Y
pearson_corr, _ = pearsonr(df['X'], df['Y'])

# Step 4: Compute Spearman correlation between X and Y
spearman_corr, _ = spearmanr(df['X'], df['Y'])

# Step 5: Print the results
print(f"Pearson Correlation (X vs Y): {pearson_corr:.2f}")
print(f"Spearman Correlation (X vs Y): {spearman_corr:.2f}")

# Optional: Visualize the data using a scatter plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['X'], df['Y'], alpha=0.6, color='blue', label="X vs Y")
plt.title("Scatter Plot of X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
